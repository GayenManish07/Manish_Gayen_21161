```python
import matplotlib.pyplot as plt
```

```python

with open('accuracies/alexnet.txt', 'r') as file:
    lines = file.readlines()

epochs = []
train_accuracy = []
test_accuracy = []

for line in lines[1:]:  # Skip the header line
    epoch, train_acc, test_acc = line.strip().split('\t')
    epochs.append(int(epoch))
    train_accuracy.append(float(train_acc))
    test_accuracy.append(float(test_acc))

plt.plot(epochs, train_accuracy, label='Train Accuracy')
plt.plot(epochs, test_accuracy, label='Test Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.title('Train and Test Accuracy over Epochs (AlexNet)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import os

# Function to read the last epoch data from a file
def read_last_epoch_data(file_path):
    with open(file_path, 'r') as file:
        lines = file.readlines()
        last_line = lines[-1].strip().split('\t')
        epoch = int(last_line[0])
        train_accuracy = float(last_line[1])
        test_accuracy = float(last_line[2])
    
    return train_accuracy, test_accuracy

# Directory containing the text files
directory = r'accuracies'

# Initialize dictionaries to store train and test accuracies for the last epoch from each file
last_epoch_train_accuracy = {}
last_epoch_test_accuracy = {}

# Iterate over each file in the directory
for filename in os.listdir(directory):
    if filename.endswith('.txt'):
        file_path = os.path.join(directory, filename)
        train_accuracy, test_accuracy = read_last_epoch_data(file_path)
        # Remove the ".txt" part from the filename
        filename_without_extension = os.path.splitext(filename)[0]
        last_epoch_train_accuracy[filename_without_extension] = train_accuracy
        last_epoch_test_accuracy[filename_without_extension] = test_accuracy

# Plotting
x = range(len(last_epoch_train_accuracy))
plt.bar(x, last_epoch_train_accuracy.values(), width=0.4, align='center', label='Train Accuracy', alpha=0.5)
plt.bar(x, last_epoch_test_accuracy.values(), width=0.4, align='edge', label='Test Accuracy', alpha=0.5)
plt.xlabel('Model')
plt.ylabel('Accuracy')
plt.title('Train and Test Accuracy after 10 epochs')
plt.xticks(x, last_epoch_train_accuracy.keys(), rotation=45)
# Change the legend location to upper right
plt.legend(loc='lower left')
plt.grid(True)
plt.show()
```
